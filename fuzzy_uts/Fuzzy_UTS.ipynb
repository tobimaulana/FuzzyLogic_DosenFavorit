```python
#import Library

import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
def RangeSubjective (_low, _high, _step):
    Subjective = np.arange(_low, _high, _step)
    return Subjective


#Pendefinisan Method Untuk Grafik Praktikum

def AlgoritmaFuzzyDosen(_rule, _range_subjective, _title):
    lo = fuzz.trimf(_range_subjective,_rule[0])
    mi = fuzz.trimf(_range_subjective,_rule[1])
    hi = fuzz.trimf(_range_subjective,_rule[2])
    
    fig, ax = plt.subplots(nrows=1, figsize=(6, 3))
    ax.plot(_range_subjective, lo, 'b', linewidth=1.5, label = 'low')
    ax.plot(_range_subjective, mi, 'g', linewidth=1.5, label = 'middle')
    ax.plot(_range_subjective, hi, 'r', linewidth=1.5, label = 'high')
    
    ax.set_title(_title)
    ax.legend()
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    
    plt.tight_layout()
    plt.show()
    
    return lo, mi, hi


#Pendefinisan Method Untuk Grafik Kedisiplinan

def AlgoritmaFuzzyDosen2(_rule, _range_subjective, _title):
    lo = fuzz.trimf(_range_subjective,_rule[0])
    mi = fuzz.trimf(_range_subjective,_rule[1])
    hi = fuzz.trimf(_range_subjective,_rule[2])
    
    fig, ax = plt.subplots(nrows=1, figsize=(6, 3))
    ax.plot(_range_subjective, lo, 'b', linewidth=1.5, label = 'low')
    ax.plot(_range_subjective, mi, 'g', linewidth=1.5, label = 'middle')
    ax.plot(_range_subjective, hi, 'r', linewidth=1.5, label = 'high')
    
    ax.set_title(_title)
    ax.legend()
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    
    plt.tight_layout()
    plt.show()
    
    return lo, mi, hi

def RangeSubjective (_low, _high, _step):
    Subjective = np.arange(_low, _high, _step)
    return Subjective


#Pendefinisan Method Untuk Grafik Metode Pembelajaran

def AlgoritmaFuzzyDosen3(_rule, _range_subjective, _title):
    lo = fuzz.trimf(_range_subjective,_rule[0])
    mi = fuzz.trimf(_range_subjective,_rule[1])
    hi = fuzz.trimf(_range_subjective,_rule[2])
    
    fig, ax = plt.subplots(nrows=1, figsize=(6, 3))
    ax.plot(_range_subjective, lo, 'b', linewidth=1.5, label = 'low')
    ax.plot(_range_subjective, mi, 'g', linewidth=1.5, label = 'middle')
    ax.plot(_range_subjective, hi, 'r', linewidth=1.5, label = 'high')
    
    ax.set_title(_title)
    ax.legend()
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    
    plt.tight_layout()
    plt.show()
    
    return lo, mi, hi


#Pendefinisan Method Untuk Grafik Keterlambatan

def AlgoritmaFuzzyDosen4(_rule, _range_subjective, _title):
    lo = fuzz.trimf(_range_subjective,_rule[0])
    mi = fuzz.trimf(_range_subjective,_rule[1])
    hi = fuzz.trimf(_range_subjective,_rule[2])
    
    fig, ax = plt.subplots(nrows=1, figsize=(6, 3))
    ax.plot(_range_subjective, lo, 'b', linewidth=1.5, label = 'low')
    ax.plot(_range_subjective, mi, 'g', linewidth=1.5, label = 'middle')
    ax.plot(_range_subjective, hi, 'r', linewidth=1.5, label = 'high')
    
    ax.set_title(_title)
    ax.legend()
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    
    plt.tight_layout()
    plt.show()
    
    return lo, mi, hi
```

```python
#Menambahkan Range Pada Grafik Praktikum

xpraktikum = RangeSubjective(10, 100, 1)
rpraktikum = np.array([
    [0, 10, 45],
    [35, 50, 80],
    [50, 100, 100]
])


#Menambahkan Range Pada Grafik Kedisiplinan

xdisiplin = RangeSubjective(10, 100, 1)
rdisiplin = np.array([
    [0, 10, 50],
    [20, 50, 90],
    [50, 100, 100]
])

#Menambahkan Range Pada Grafik Metode Pembelajaran

xmtdbelajar = RangeSubjective(10, 100, 1)
rmtdbelajar = np.array([
    [0, 5, 50],
    [30, 50, 80],
    [45, 100, 100]
])

#Menambahkan Range Pada Grafik Keterlambatan

xtolelambat = RangeSubjective(10, 100, 1)
rtolelambat = np.array([
    [0, 10, 50],
    [25, 50, 85],
    [50, 100, 100]
])


#Menampilkan Grafik
lo_praktikum, mi_praktikum, hi_praktikum = AlgoritmaFuzzyDosen(rpraktikum, xpraktikum, 'Praktikum')
lo_disiplin, mi_disiplin, hi_disiplin = AlgoritmaFuzzyDosen2(rdisiplin, xdisiplin, 'Kedisiplinan')
lo_mtdbelajar, mi_mtdbelajar, hi_mtdbelajar = AlgoritmaFuzzyDosen3(rmtdbelajar, xmtdbelajar, 'Metode Pembelajaran')
lo_tolelambat, mi_tolelambat, hi_tolelambat = AlgoritmaFuzzyDosen4(rtolelambat, xtolelambat, 'Keterlambatan')
```

```python
#Pendefinisan Method Untuk Fungsi Keanggotaan Grafik Praktikum

def FungsiKeanggotaan(_range, _lo, _mi, _hi, _nilai):
    lo = fuzz.interp_membership(_range, _lo, _nilai)
    mi = fuzz.interp_membership(_range, _mi, _nilai)
    hi = fuzz.interp_membership(_range, _hi, _nilai)
    
    return lo, mi, hi

#Pendefinisan Method Untuk Fungsi Keanggotaan Grafik Kedisiplinan

def FungsiKeanggotaan2(_range, _lo, _mi, _hi, _nilai):
    lo = fuzz.interp_membership(_range, _lo, _nilai)
    mi = fuzz.interp_membership(_range, _mi, _nilai)
    hi = fuzz.interp_membership(_range, _hi, _nilai)
    
    return lo, mi, hi

#Pendefinisan Method Untuk Fungsi Keanggotaan Grafik Metode Pembelajaran

def FungsiKeanggotaan3(_range, _lo, _mi, _hi, _nilai):
    lo = fuzz.interp_membership(_range, _lo, _nilai)
    mi = fuzz.interp_membership(_range, _mi, _nilai)
    hi = fuzz.interp_membership(_range, _hi, _nilai)
    
    return lo, mi, hi

#Pendefinisan Method Untuk Fungsi Keanggotaan Grafik Keterlambatan

def FungsiKeanggotaan4(_range, _lo, _mi, _hi, _nilai):
    lo = fuzz.interp_membership(_range, _lo, _nilai)
    mi = fuzz.interp_membership(_range, _mi, _nilai)
    hi = fuzz.interp_membership(_range, _hi, _nilai)
    
    return lo, mi, hi
```

```python
#Pendefinisian Status Keanggotaan

def Status(_keanggotaan):
    status = ""
    if _keanggotaan[0] > _keanggotaan[1] and _keanggotaan[0] > _keanggotaan[2]:
        status = "low"
    elif _keanggotaan[1] > _keanggotaan[0] and _keanggotaan[1] > _keanggotaan[2]:
        status = "middle"
    elif _keanggotaan[2] > _keanggotaan[0] and _keanggotaan[2] > _keanggotaan[1]:
        status = "high"
        
    return status
```
